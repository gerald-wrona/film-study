```python
import pandas as pd
films = pd.read_csv('films.csv')
```

```python
films
```

```python
written = films.groupby(['Writer*']).size().reset_index(name='total_written')
```

```python
films = films.merge(written, on='Writer*', how="left")
```

```python
directed = films.groupby(['Director']).size().reset_index(name='total_directed')
films = films.merge(directed, on='Director', how="left")
```

```python
languages = films.groupby(['Language']).size().reset_index(name='total_language')
films = films.merge(languages, on='Language', how="left")
```

```python
films['Writer*'].value_counts()
writer_chart = films.drop(films[films.total_directed < 2].index)
writer_chart['Writer*'].hist()
```

```python
films.groupby(['Writer*']).size().reset_index(name='total_written').sort_values(by='total_written', ascending=False)
```
